```python
# Importing the imporant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Reading the Dataset
df = pd.read_csv('StudentsPerformance.csv')

# Displaying first five rows
df.head()
```

```python
# Exploring the dataset
print("Shape of dataset:", df.shape)
print("\nColumn Names:\n", df.columns.tolist(),sep='')
print("\nData Types:\n", df.dtypes,sep='')
print("\nMissing Values:\n", df.isnull().sum(),sep='')
```

```python
# Basic statistics for all numeric columns
df.describe()
```

```python
# visualizing the distribution of math scores using histograph
plt.figure(figsize=(8,5))
plt.hist(df['math score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Percentages')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Visualising the comaparison of all 3 scores by histograph
plt.figure(figsize=(10,6))
plt.hist(df['math score'], bins=20, color='skyblue', edgecolor='black', alpha=0.5, label='Math')
plt.hist(df['reading score'], bins=20, color='salmon', edgecolor='black', alpha=0.5, label='Reading')
plt.hist(df['writing score'], bins=20, color='lightgreen', edgecolor='black', alpha=0.5, label='Writing')
plt.title('Comparison of All 3 scores')
plt.xlabel('Percentages')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

```python
# Average scores using NumPy
math_average = np.mean(df['math score'])
reading_average = np.mean(df['reading score'])
writing_average = np.mean(df['writing score'])

print("Average Math Score      :", round(math_average, 2))
print("Average Reading Score   :", round(reading_average, 2))
print("Average writing Score   :", round(writing_average, 2))
```

```python
# Bar chart of average scores
sub=['Math','Reading','Writing']
avg=[math_average, reading_average, writing_average]
colr=['skyblue', 'salmon', 'lightgreen']
plt.figure(figsize=(8,5))
plt.bar(sub, avg, color=colr, edgecolor='black')
plt.title('Average Score by Subject')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.ylim(60,72)
plt.show()
```

```python
# Grouping students by test preparation course
test_prep = df.groupby('test preparation course')['math score'].mean()
print(test_prep)
```

```python
# Bar chart of test prep vs no test prep
test_prep = df.groupby('test preparation course')['math score'].mean()
plt.figure(figsize=(8,6))
plt.bar(test_prep.index, test_prep.values, color=['lightgreen', 'salmon'], alpha=0.5, edgecolor='black')
plt.title('Graph based on test preparation coures on math')
plt.xlabel('Test preparation Coures')
plt.ylabel('Average maths score')
plt.ylim(60,75)
plt.show()
```

```python
# Average scores by gender
gendr_score = df.groupby('gender')[['math score', 'writing score', 'reading score']].mean()
print(gendr_score)
```

```python
# Gender comparison graph
gendr_score.plot(kind='bar', figsize=(8,10), color=['lightgreen', 'salmon', 'skyblue'], edgecolor='black')
plt.title('Average scores by gender')
plt.xlabel("Gender")
plt.ylabel("Average Scores")
plt.xticks(rotation=0)
plt.legend(['Maths Score', 'Writing Score', 'Reading Score'])
plt.ylim(50,80)
///plt.show()
```

```python
# Parental education impact
parent_edu = df.groupby('parental level of education')['math score'].mean()
print(parent_edu)
```

```python
# Parental education impact graph
parent_edu.plot(kind='bar', figsize=(8,10), color="skyblue", edgecolor='black')
plt.title('Average scores by parantal level of education')
plt.xlabel('Parantal level of education')
plt.ylabel('Average scores')
plt.xticks(rotation=45)
plt.ylim(50,75)
plt.show()
```
